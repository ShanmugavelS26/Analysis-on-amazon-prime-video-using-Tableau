Handle missing values in specific columns

```python
df1['director']= df1["director"].replace(np.nan,'unknown')
```

```python
df1['cast']= df1['cast'].replace(np.nan,"unknown")
```

```python
df1['rating']=df1["rating"].fillna(df1['rating'].mode()[0])
```

```python
df1['date_added']= df['date_added'].ffill()
```
